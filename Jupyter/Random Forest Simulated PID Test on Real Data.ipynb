```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
import csv
import pdb
import os
import matplotlib.pyplot as plt
import timeit
```

```python
def load_data(directory, target_var):
    full_data = []
    if directory[-1] != '/':
        directory = directory + '/'
    for root,dir,files in os.walk(directory):
        for file in files:
            if file.endswith(".csv"):
                filepath = directory + file
                print('Loading: ', file)
                csv_data = np.genfromtxt(filepath, delimiter=',')
                full_data.append(csv_data)
    try:
        features = np.vstack(full_data)
        ground_truth = features[:,target_var]
        features = np.delete(features, target_var, axis=1)
    except:
        print("Loading Error: Data not found")
    
    return features, ground_truth, full_data
    
```

```python
truth_index = 2
start = timeit.default_timer()
training_features, training_truth, _ = load_data('/mnt/d/git/heat_chamber_code/jupyter/pid train/', truth_index)
```

```python
forest = RandomForestRegressor(random_state=42)
forest.fit(training_features, training_truth)
```

```python
test_features, test_truth, run_list = load_data('/mnt/d/git/heat_chamber_code/jupyter/real_cross_validation/', truth_index)
prediction_array = []
truth_array = []
score = []
for run in range(len(run_list)):
    run_predictions = []
    run_truth = []
    run_features = []
    for row in run_list[run]:
        run_truth.append(row[truth_index])
        row = np.delete(row, truth_index)
        prediction = forest.predict([row])
        run_features.append(row)
        run_predictions.append(prediction)
    run_predictions = np.vstack(run_predictions)
    run_truth = np.vstack(run_truth)
    score.append(forest.score(run_features, run_truth))
    prediction_array.append(run_predictions)
    truth_array.append(run_truth)
```

```python
for run in range(len(run_list)):
    plt.plot(range(len(truth_array[run])), prediction_array[run], label="Prediction")
    plt.plot(range(len(truth_array[run])), truth_array[run], label="Ground truth")
    plt.legend(loc="lower right")
    plt.show()
    print("Score: ", score[run])
    stop = timeit.default_timer()
    print('Time: ', stop - start)  
```
